```python
import numpy as np
import matplotlib.pyplot as plt
bins=np.linspace(df['Cholesterol'].min(),df['Cholesterol'].max(),4)
groups=['low','medium','high']
df['Cholesterol bined']=pd.cut(df['Cholesterol'],bins,groups,include_lowest=True)
# Count how many values fall into each bin
bin_counts = df['Cholesterol bined'].value_counts().sort_index()

# Plot bar chart
bin_counts.plot(kind='bar', color='lightgreen')
plt.xlabel('Category')
plt.ylabel('Number of Samples')
plt.title('Number of Samples in Each Bin')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
df.drop('Cholesterol bined',axis=1,inplace=True)
```
